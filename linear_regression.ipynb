```python
import pandas as pd
```

```python
df=pd.read_csv("df_test.csv")
```

```python
df.head(10)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
df['day']=df['date'].dt.day
```

```python
df.head(10)
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum
```

```python
df.dropna(inplace=True)
```

```python
df['has_basement']=df['has_basement'].map({False:0,True:1})
df['renovated']=df['renovated'].map({False:0,True:1})
df['nice_view']=df['nice_view'].map({False:0,True:1})
df['perfect_condition']=df['perfect_condition'].map({False:0,True:1})
df['has_lavatory']=df['has_lavatory'].map({False:0,True:1})
df['single_floor']=df['single_floor'].map({False:0,True:1})
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
pip install -U scikit-learn
```

```python
x=df.drop('price',axis=1)
y=df['price']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_pred,y_test)
```
